```python
import requests
import pandas
```

```python
abbr = "NY"
```

```python
st_summary = requests.get('https://colorado.rstudio.com/rsc/content/2280/summary?state=' + abbr)
```

```python
st_summary.text
```

```python
pandas.DataFrame.from_dict(st_summary.json())
```

```python
st_detail = requests.get('https://colorado.rstudio.com/rsc/content/2280/state?state=' + abbr)
```

```python
counties = pandas.DataFrame.from_dict(st_detail.json())
```

```python
counties.nlargest(10, 'population')
```

counties.nlargest(10, 'hospitals')

```python
counties[counties.result == -1]
```

```python
counties[counties.result == 1]
```
