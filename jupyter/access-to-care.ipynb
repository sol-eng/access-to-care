```python
import requests
import pandas
```

```python
abbr = "NY"
```

```python
st_summary = requests.get('https://colorado.rstudio.com/rsc/access-to-care/api/summary?state=' + abbr)
```

```python
st_summary.text
```

```python
pandas.DataFrame.from_dict(st_summary.json())
```

```python
st_detail = requests.get('https://colorado.rstudio.com/rsc/access-to-care/api/state?state=' + abbr)
```

```python
counties = pandas.DataFrame.from_dict(st_detail.json())
```

```python
counties.nlargest(10, 'population')
```
